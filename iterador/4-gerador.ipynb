Yield
- É uma palavra chave para criar geradores. Quando você usa ele em uma função, você transforma a função em um gerador, que é um tipo especial de um iterador.
- Return: Quando é usado numa função, essa é encerrada e o valor é retornado. Chamadas subsequentes à função iniciam a execução novamente do código.
- Yield: Pode ser chamada novamente e retormar a execução do ponto onde foi pausada, permitindo a produção de uma sequência de valores ao longo do tempo

```python
def numeros_impares(n):
    for i in range(1, n+1, 2):
        yield i
```

```python
for numero in numeros_impares(10):
    print(numero)
```
